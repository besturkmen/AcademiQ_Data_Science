Standardizasyon

```python
dt.columns = (dt.columns
              .str.strip()
              .str.lower()
              .str.replace(" ","_")
              .str.replace("-","_")
              )
print("\nTemizlenmiş Sütun İsimleri:")
#print(dt.columns.tolist())
print("\n".join(dt.columns))
```

```python
string_cols = dt.select_dtypes(include="object").columns
for x in string_cols:
  dt[x] =dt[x].astype(str).str.strip() #astype(str) = sütundaki verileri string'e dönüştür. ///
# .str = sadece string olan sütunlarda strip işlemini uygulamak için yazıldı.
# strip() = baştaki ve sondaki boşlukları temizler.
```

```python
critical_cols = ["repository_name","stars_count"]
dt = dt.dropna(subset=critical_cols)
```

```python
numeric_cols = ["stars_count","forks_count","watchers_count",
"open_issues_count"]

for col in numeric_cols:
  if col in dt.columns:
    dt[col] = pd.to_numeric(dt[col], errors="coerce")

    # errors=coerce ifadesi sayıya çevirilemeyecek bir ifadeyi NaN şeklinde ifade eder.  "100" = 100 //// "abc" = NaN /// "50.5" = 50.5

    #NaN = Not a Number ( numpy ile gelir. Kullanmak için numpy'a ihtiyaç vardır.)
    #NA = Not Available (bu veri yok.)(pandas ile gelir.)
```
